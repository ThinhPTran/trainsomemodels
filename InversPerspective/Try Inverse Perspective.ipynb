```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMAGE_H = 223
IMAGE_W = 1280
```

```python
src = np.float32([[0, IMAGE_H], [1207, IMAGE_H], [0, 0], [IMAGE_W, 0]])
dst = np.float32([[569, IMAGE_H], [711, IMAGE_H], [0, 0], [IMAGE_W, 0]])
M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
Minv = cv2.getPerspectiveTransform(dst, src) # Inverse transformation
print(M)
```

```python
img = cv2.imread('../Image/Road_inverse_perspective.jpg') # Read the test img
crop_img = img[450:(450+IMAGE_H), 0:IMAGE_W] # Apply np slicing for ROI crop
warped_img = cv2.warpPerspective(crop_img, M, (IMAGE_W, IMAGE_H)) # Image warping
```

```python
plt.imshow(cv2.cvtColor(warped_img, cv2.COLOR_BGR2RGB)) # Show results
```

```python
plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB)) # Show results
```
