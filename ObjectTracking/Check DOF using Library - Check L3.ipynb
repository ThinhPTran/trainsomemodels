```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image = np.fromfile("../Image/dofoutput_mode_no_fusion/outputimage_dof_l3.bin", dtype=np.uint16)
# image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
cv2.imwrite("../Image/Output_Lib_L3_204_960.png", image.reshape((204,960)))
```

```python
image_L3_204_960 = cv2.imread("../Image/Output_Lib_L3_204_960.png")
image_L3_204_960_gray = cv2.cvtColor(image_L3_204_960, cv2.COLOR_BGR2GRAY)
```

```python
Vx = np.zeros((204,480))
Vy = np.zeros((204,480))
```

```python
for i in range(204):
    for j in range(480):
        Vx[i, j] = image_L3_204_960_gray[i, 2*j+0]
        Vy[i, j] = image_L3_204_960_gray[i, 2*j+1]
```

```python
f, axarr = plt.subplots(1,1, figsize=(20, 20))
axarr.imshow(image_L3_204_960_gray)
```

```python
f, axarr = plt.subplots(1,1, figsize=(20, 20))
axarr.imshow(Vx)
```

```python
f, axarr = plt.subplots(1,1, figsize=(20, 20))
axarr.imshow(Vy)
```
