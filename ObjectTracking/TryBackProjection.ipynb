```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
orig01 = cv2.imread("../Image/hand_01.jpg")
orig02 = cv2.imread("../Image/hand_02.jpg")
hsv = cv2.cvtColor(orig01, cv2.COLOR_BGR2HSV)
```

ch = (0, 0)
hue = np.empty(hsv.shape, hsv.dtype)

```python
cv2.mixChannels([hsv], [hue], ch)
```

```python
plt.imshow(hsv)
```

```python
plt.imshow(hue)
```

```python
bins = 25
histSize = max(bins, 2)
ranges = [0, 180]
hist = cv2.calcHist([hue], [0], None, [histSize], ranges, accumulate=False)
```

```python
cv2.normalize(hist, hist, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
```

```python
backproj = cv2.calcBackProject([hue], [0], hist, ranges, scale=1)
```

```python
plt.imshow(backproj, cmap='gray')
```
