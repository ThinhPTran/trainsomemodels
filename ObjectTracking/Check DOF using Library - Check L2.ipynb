```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image = np.fromfile("../Image/dofoutput_mode_no_fusion/outputimage_dof_l2.bin", dtype=np.uint16)
cv2.imwrite("../Image/Output_Lib_L2_408_1920.png", image.reshape((408,1920)))
```

```python
image_L2_408_1920 = cv2.imread("../Image/Output_Lib_L2_408_1920.png")
image_L2_408_1920_gray = cv2.cvtColor(image_L2_408_1920, cv2.COLOR_BGR2GRAY)
```

```python
f, axarr = plt.subplots(1,1, figsize=(20, 20))
axarr.imshow(image_L2_408_1920_gray)
```
