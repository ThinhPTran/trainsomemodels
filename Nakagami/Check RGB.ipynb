```python
def get_max_rgb(imgin):
    inmap = imgin.load()
    out = [0, 0, 0]
    for i in range(imgin.size[0]):
        for j in range(imgin.size[1]):
            if (inmap[i,j][0]>out[0]):
                out[0] = inmap[i,j][0]
            if (inmap[i,j][1]>out[1]):
                out[1] = inmap[i,j][1]
            if (inmap[i,j][2]>out[2]):
                out[2] = inmap[i,j][2]
    return out
```

```python
get_max_rgb(img_car1)
```

```python
get_max_rgb(img_car2)
```

```python
plt.imshow(img_car1)
```

```python
plt.imshow(img_car2)
```

```python
# Remove green and blue
def remove_green_blue(imgin):
    inmap = imgin.load()
    out = [0, 0, 0]
    for i in range(imgin.size[0]):
        for j in range(imgin.size[1]):
            inmap[i,j]= (inmap[i,j][0],0,0)
    return out
```

```python
remove_green_blue(img_car1)
plt.imshow(img_car1)
```

```python
remove_green_blue(img_car2)
plt.imshow(img_car2)
```
