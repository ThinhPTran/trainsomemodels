```python
import PIL as pil
from matplotlib import pyplot as plt
from skimage.color import rgb2hsv
import numpy as np
```

```python
img_car1 = pil.Image.open("../Image/car_no_light_1280x720.jpg")
img_car2 = pil.Image.open("../Image/car_with_brake_light_1280x720.jpg")
```

```python
#Convert to HSV
img_car1_map = img_car1.load()
img_car2_map = img_car2.load()
```

```python
img_car1_map[0,0]
```

```python
hsv_img = img_car1.convert(mode='HSV')
```

```python
hsv_img_map = hsv_img.load()
hsv_img_map[0,0]
```

```python
# The following example converts an RGB image 
# (linearly calibrated according to ITU-R 709, 
# using the D65 luminant) to the CIE XYZ color space:
rgb2xyz = (
    0.412453, 0.357580, 0.180423, 0,
    0.212671, 0.715160, 0.072169, 0,
    0.019334, 0.119193, 0.950227, 0)
out = img_car1.convert("RGB", rgb2xyz)
```

```python
out_map = out.load()
out_map[0,0]
```
