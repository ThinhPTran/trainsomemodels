```python
from scipy.stats import nakagami
import matplotlib.pyplot as plt
import numpy as np
```

```python
nu = 4.97
mean, var, skew, kurt = nakagami.stats(nu, moments='mvsk')
```

```python
print("mean: " + str(mean) + " var: " + str(var) + " skew: " + str(skew) + " kurt: " + str(kurt))
```

```python
x = np.linspace(nakagami.ppf(0.01, nu), nakagami.ppf(0.99, nu), 100)
rv = nakagami(nu)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
ax.plot(x, nakagami.pdf(x, nu), 'r-', lw=10, alpha=0.6, label='nakagami pdf')
ax.plot(x, nakagami.cdf(x, nu), 'b-', lw=2, label='nakagami cdf')
ax.legend(['frozen pdf', 'nakagami pdf'])
```

```python
vals = nakagami.ppf([0.001, 0.5, 0.999], nu)
np.allclose([0.001, 0.5, 0.999], nakagami.cdf(vals, nu))
```

```python
# Generate random in nakagami distribution
r = nakagami.rvs(nu, size=1000)
```

```python
#Check histogram
fig, ax = plt.subplots(1,1)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
ax.plot(x, nakagami.pdf(x, nu), 'r-', lw=10, alpha=0.6, label='nakagami pdf')
ax.plot(x, nakagami.cdf(x, nu), 'b-', lw=2, label='nakagami cdf')
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
```

```python
# Check m
r2 = np.multiply(r, r)
r4 = np.multiply(r2, r2)
```

```python
np.average(r2)*np.average(r2)
```

```python
# Use np.var and use my calculation
# Var(X) = E(X^2) - E(X)^2
np.var(r2)
```

```python
np.average(r4) - np.average(r2)*np.average(r2)
```
